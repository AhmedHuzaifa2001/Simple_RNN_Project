```python
from tensorflow.keras.preprocessing.text import one_hot # type: ignore
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocubulary size
voc_size = 10000
```

```python
one_hot_repr = [one_hot(words , voc_size)for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
import numpy as np
from tensorflow.keras.models import Sequential
```

```python
sent_size = 8
```

```python
embedding_docs = pad_sequences(one_hot_repr , padding = 'pre' , maxlen = sent_size)
```

```python
embedding_docs
```

```python
## Feature Representation
dim = 10 ## How many Features
```

```python
model = Sequential()
model.add(Embedding(voc_size , dim , input_length = sent_size))
model.compile('adam' , 'mse')
```

```python
model.summary()
```

```python
## Convert to Vectors Each word is having 10 features

model.predict(embedding_docs)
```

```python
## First Sentence
model.predict(embedding_docs[0])
```
